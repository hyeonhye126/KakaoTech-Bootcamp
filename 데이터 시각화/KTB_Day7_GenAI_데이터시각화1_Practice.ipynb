```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
# 데이터 정의
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [150, 200, 180, 220, 250, 230]
```

```python
# 막대 그래프 생성
plt.figure(figsize=(10, 6)) # 그래프 크기 설정
plt.bar(months, sales, color='pink', label='Sales')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 데이터 정의
labels = ['Product A', 'Product B', 'Product C', 'Product D']
sizes = [400, 300, 500, 100]
colors1 = ['#ff9999', '#ffb3b3', '#ffcccc', '#ffe6e6'] # 그라데이션 - 진한 빨강, 중간 빨강, 연한 빨강, 매우 연한 빨강
colors2 = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)
```

```python
# 간결한 디자인의 파이차트
plt.figure(figsize=(8, 8)) # 그래프 크기 설정
plt.pie(sizes, colors=colors1, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Sales Distribution - Simple Design')
plt.axis('equal')
plt.show()

# autopct: 각 조각에 표시될 비율의 형식을 지정한다. `'%1.1f%%'`는 소수점 첫째 자리까지 비율을 표시하고, 뒤에 `%` 기호를 붙이겠다는 의미
# startangle :파이차트의 시작 각도를 지정한다. 기본값은 0도이지만, 이 예제에서는 140도로 설정되어 있어서 파이차트가 시계 방향으로 140도 회전된 상태에서 시작한다.
```

```python
# 복잡한 디자인의 파이차트
plt.figure(figsize=(8,8)) # 그래프 크기 설정
plt.pie(sizes, explode=explode, labels=labels, colors=colors2, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Sales Distribution - Complex Design')
plt.axis('equal') # 파이차트를 원형으로 유지
plt.show()
```

```python
# 데이터 정의
data = {'Year': [2018, 2019, 2020, 2021], 'Visitors': [12000, 18000, 23000, 30000]}
df = pd.DataFrame(data)
```

```python
# 선그래프 생성
plt.figure(figsize=(10,6)) # 그래프 크기 설정
sns.lineplot(x='Year', y='Visitors', data=df, marker='*', label='Visitors')
plt.title('Annual Website Visitors') # 제목 추가
plt.xlabel('Year') # x축 레이블 추가
plt.ylabel('Number of Visitors') # y축 레이블 추가
plt.legend() # 범례 추가
plt.grid(True) # 그리드 추가
plt.show()
```

```python
# 데이터 정의
data = {'Year': [2018, 2019, 2020, 2021],
        'Product A': [1500, 2000, 2400, 2800],
        'Product B': [2300, 2700, 3000, 3500],
        'Product C': [1800, 2200, 2600, 3000]}
df = pd.DataFrame(data)
```

```python
# 복합 차트 형성
plt.figure(figsize=(10,6)) # 그래프 크기 설정
sns.lineplot(x='Year', y='Product A', data=df, marker='o',label='Product A')
sns.lineplot(x='Year', y='Product B', data=df, marker='o',label='Product B')
sns.lineplot(x='Year', y='Product C', data=df, marker='o',label='Product C')
plt.title('Annual Product Sales') # 제목 추가
plt.xlabel('Year') # x축 레이블 추가
plt.ylabel('Sales') # y축 레이블 추가
plt.legend() # 범례 추가
plt.grid(True) # 그리드 추가
plt.show()
```

```python
# 데이터 생성
np.random.seed(10)
data = np.random.randn(1000) # 표준 정규분포를 따르는 1000개의 랜덤 데이터를 생성
```

```python
# 히스토그램 생성
plt.figure(figsize=(10, 6)) # 그래프 크기 설정
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Random Data') # 제목 추가
plt.xlabel('Value') # x축 레이블 추가
plt.ylabel('Frequency') # y축 레이블 추가
plt.grid(True) # 그리드 추가
plt.show
```

```python
# 데이터 생성
np.random.seed(10)
data = {
    'x': np.random.randn(100),
    'y': np.random.randn(100)
}
df = pd.DataFrame(data)
```

```python
# 산점도 생성
plt.figure(figsize=(10, 6)) # 그래프 크기 설정
sns.scatterplot(x='x', y='y', data=df)
plt.title('Scatter Plot of Random Data') # 제목 추가
plt.xlabel('X') # x축 레이블 추가
plt.ylabel('Y') # y축 레이블 추가
plt.grid(True) # 그리드 추가
plt.show()
```

```python
# 데이터 생성
np.random.seed(10)
data = {
    'Category': np.random.choice(['A','B','C'], 100), # ['A', 'B', 'C']에서 무작위로 100개의 요소를 추출
    'Value': np.random.randn(100)
}
df = pd.DataFrame(data)
```

```python
# 박스플롯 생성
plt.figure(figsize=(10, 6)) # 그래프 크기 설정
sns.boxplot(x='Category', y='Value', data=df, hue='Category', palette=['#FF9999', '#66B3FF', '#99FF99'])
plt.title('Box Plot of Random Data by Category') # 제목 추가
plt.xlabel('Category') # x축 레이블 추가
plt.ylabel('Value') # y축 레이블 추가
plt.grid(True) # 그리드 추가
plt.show()
```
